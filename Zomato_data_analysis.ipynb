```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Zomato.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)
```

```python
df['rate']=df['rate'].apply(handleRate)
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type of restaurant')
```

```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,c = 'red',marker='o')
plt.xlabel('Type of restaurant',c='blue',size = 20)
plt.ylabel("Votes",c = 'blue', size = 20)
```

```python
df.head()
```

```python
plt.hist(x=df['rate'],bins = 5)
plt.title("Ratins Distibution")
plt.show()
```

```python
avg_spending = df['approx_cost(for two people)']
sns.countplot(x=avg_spending)
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data=df)
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', values='rate', aggfunc='size', fill_value = 0)
sns.heatmap(pivot_table,annot = True,cmap = 'YlGnBu', fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listen In (Type)")
plt.show()
```
